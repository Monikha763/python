```python
import pandas as pd

df = pd.read_csv('netflix_titles_nov_2019.csv')
df['date_added'] = pd.to_datetime(df['date_added'])
df
```

```python
df["year"]=df["date_added"].dt.year
df["month"]=df["date_added"].dt.month
df
```

```python
df['value']=df['duration'].str.extract(r'(\d+\.?\d*)')
df['unit']=df['duration'].str.extract(r'([a-zA-Z]+)')
display(df[['duration','value','unit']])
```

```python
print(df['type'].value_counts())
```

```python
top_5_countries=df["country"].value_counts().head(5)
print(top_5_countries)
```

```python
titles_per_year = df['year_added'].value_counts()
print(titles_per_year)
```

```python
ur = df['rating'].unique()
print("Unique ratings:", ur)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=df,x="type")
plt.title("count of Movies vs TV Shows")
plt.xlabel("type")
plt.ylabel("count")
plt.show()
```

```python
sns.lineplot(x=titles_per_year.index,y=titles_per_year.values)
plt.title("number of titles added per year")
plt.xlabel("year")
plt.ylabel("number of titles")
plt.show()
```
